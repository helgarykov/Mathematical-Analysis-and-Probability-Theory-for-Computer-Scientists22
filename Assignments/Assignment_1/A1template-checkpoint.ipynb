```python
# Allows viewing figures inline in the notebook
# If you are using Jupyter-lab, you may want to replace "inline" by "widget"
%matplotlib inline
# numpy is a library for numerical computation
import numpy as np
# matplotlib is a plotting library
import matplotlib.pyplot as plt
import matplotlib
# Sets larger size for viewed figures
matplotlib.rcParams['figure.figsize'] = (20,8)
```

### Exercise 2
Please use the cell below for the plot in Exercise 2i)

```python
x = np.linspace(-5, 5 ,100)  # define x-value interval[a,b] og antallet af punkter (grønne)
x[:50] = np.zeros(50)        #x[:50] svarer til c# x[..50] 
x1 = np.linspace(-5, 5, 100) #x1-blå de 10 første punkter i array er sat til -4
x1[:10] = [-4]
x1[50:] = np.zeros(50)
with plt.style.context('Solarize_Light2'): # plot style 
    
    fig, ax = plt.subplots() #matplotlib.subplots (wrapper) laver 1 figur, hedder ax
    
    ax.set_xlim([-5, 5])    #hardcoder intervalet på x-aksen
    ax.hlines(y=10, xmin=-5, xmax=-4, label='10', color='red') #rød grad horizontal line for den røde graf
    ax.plot(x1, (-x1**3) /32, label='- x ^ 3 / 32', color='blue')
    ax.plot(x, (x**2) / 3, label='x ^ 2 / 3', color='green')
    
    ax.legend() #viser labels
```
